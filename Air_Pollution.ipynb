```python
import pandas as pd
import io
import numpy as np
import matplotlib.pyplot as plt
```

```python
import plotly.express as px
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv(io.StringIO(uploaded['death-rates-from-air-pollution.csv'].decode('utf-8')))
df.head()
```

```python
del df['Code']
```

```python
df.info()
```

```python
df.rename(columns={'Deaths - Air pollution - Sex: Both - Age: Age-standardized (Rate)': 'Deaths by air pollution', 
                 'Deaths - Household air pollution from solid fuels - Sex: Both - Age: Age-standardized (Rate)': 'Deaths by household air pollution from fuels' ,
                 'Deaths - Ambient particulate matter pollution - Sex: Both - Age: Age-standardized (Rate)': 'Death by ambient particulate pollution', 
                 'Deaths - Ambient ozone pollution - Sex: Both - Age: Age-standardized (Rate)': 'Deaths by ambient ozone pollution'}, inplace=True)
```

```python
Country_df = df.groupby("Entity")
Air_10 = Country_df['Deaths by air pollution'].sum().sort_values(ascending=False)[:10]
ax = Air_10.plot(kind='bar', title ="Top 10 Countries by number of Deaths by air pollution", figsize=(7,5), legend=True, fontsize=12, color='blue')
```

```python
Country_df = df.groupby("Entity")
Household_10 = Country_df['Deaths by household air pollution from fuels'].sum().sort_values(ascending=False)[:10]
ax = Household_10.plot(kind='bar', title ="Top 10 Countries by number of Deaths by household air pollution from fuels", figsize=(7,5), legend=True, fontsize=12, color='mediumturquoise')
```

```python
Country_df = df.groupby("Entity")
particulate_10 = Country_df['Death by ambient particulate pollution'].sum().sort_values(ascending=False)[:10]
ax = particulate_10.plot(kind='bar', title ="Top 10 Countries by number of Death by ambient particulate pollution", figsize=(7,5), legend=True, fontsize=12, color='mediumseagreen')
```

```python
Country_df = df.groupby("Entity")
Ozone_10 = Country_df['Deaths by ambient ozone pollution'].sum().sort_values(ascending=False)[:10]
ax = Ozone_10.plot(kind='bar', title ="Top 10 Countries by number of Deaths by ambient ozone pollution", figsize=(7,5), legend=True, fontsize=12, color='salmon')
```

```python
US = df[df['Entity'] == 'United States'].set_index(['Year'])
China = df[df['Entity'] == 'China'].set_index(['Year'])
India = df[df['Entity'] == 'India'].set_index(['Year'])
Russia = df[df['Entity'] == 'Russia'].set_index(['Year'])
Australia = df[df['Entity'] == 'Australia'].set_index(['Year'])
Canada = df[df['Entity'] == 'Canada'].set_index(['Year'])
```

```python
l = [US,China,India,Russia,Australia,Canada]
for i in l:
   i.plot(title = ' Pollution Related Deaths in ' + i.iloc[0][0], ylabel = 'Deaths' , figsize = (14,7))
```

```python
pm = pd.read_csv(io.StringIO(uploaded['PM2.5 Global Air Pollution 2010-2017.csv'].decode('utf-8')))
pm.head()
```

```python
del pm['Country Code']
```

```python
pm['Average'] = ((pm['2010'] + pm['2011'] + pm['2012']
                                     + pm['2013'] + pm['2014'] + pm['2015']
                                     + pm['2016'] + pm['2017'])/8)

pm = pm.set_index('Country Name')
```

```python
# category_dist=df.groupby(['Category']).sum()
p = pm.sort_values('Average',ascending = False)['Average'].head(10)

fig = px.pie(values=p.values, names=p.index, title='Countries with highest PM 2.5 Air Pollution (2010-2017)')

fig.write_image("image.svg")
fig.write_image("image.png")
fig.show("png")


```

```python
least_pm = pm.sort_values('Average',ascending = True)['Average'].head(10)
least_pm.plot.bar(figsize = (10,5), title = 'Countries with lowest PM 2.5 Air Pollution (2010-2017)', ylabel = "Levels of pollution", color = 'limegreen')
```
